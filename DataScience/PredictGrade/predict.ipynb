## Mean

```python
import numpy as np

df = df.replace('na', np.nan)
df = df.iloc[:, 1:]
df
```

```python
df['seal'].dtypes
```

```python
for i in df.columns:
    data = df[i]

    if pd.api.types.is_numeric_dtype(data):
        mean = data.mean()
        std = data.std()
        print(f"Mean of {i} : {mean}")
        print(f"Std of {i} : {std}")
        print("------------------------")

    else:
        for colname, coldata in df.items():
            if coldata.dtype == 'object':
                print(
                    f"""
==============================
Iterating column: {colname}"""
                )
                arrItem = []
                for item in coldata:
                    # keep only non-missing values that look like integers
                    if pd.notna(item) and str(item).isdigit():
                        arrItem.append(int(item))
                print(f"Mean from {colname} : {np.mean(arrItem)}")
                print(f"Std from {colname} : {np.std(arrItem)}")
        
```
