```python
lst = [1, 2, 3, 4, 5]
# The multiply each value by 2
doubledLst = [
    x * 2 for x in lst
]

print(doubledLst)
```

```python
import numpy as np
# Or just use an array using Numpy
arrayLst = np.array(lst)
doubledArrayLst = arrayLst * 2

print(doubledArrayLst)
```

```python
# To extract the odd numbers from a list we can just simply use iteration and extract reminder

oddNumbers = []

for x in lst:
    x % 2 != 0 and oddNumbers.append(x)

print(oddNumbers)
```
